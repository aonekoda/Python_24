# Pandas로 시각화

Pandas에서 시리즈, 데이터프레임 객체를 사용하여 간단한 시각화가 가능하다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Data Loading

iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")
```

### pandas를 이용한 히스토그램 그리기

```python
iris['sepal_width'].hist(bins=10, grid=False)

plt.show()
```

### 여러개 그룹의 히스토그램

```python
# pandas 사용하기
iris.groupby('species')['petal_length'].hist(bins=7, grid=False, alpha=0.5)
plt.legend(['setosa','versicolor','virginica'])
```

### 커널 밀도

```python
fig, ax = plt.subplots() 
ax = tips['tip'].plot.kde()
```

###  pandas 로 막대 그래프 그리기

```python
# 요일별 tip 평균
tips_mean_by_day = tips.groupby('day').tip.mean().sort_values(ascending=False)

tips_mean_by_day.plot(kind='bar',x='day',y='tip')
plt.show()
```

```python
df = pd.crosstab(tips.day, tips.sex)
df
```

```python
df.plot(kind='bar')
```

### Pandas로 산점도 그리기

```python
groups = iris.groupby('species')
ax = groups.get_group('setosa').plot.scatter(x='petal_length', y='petal_width', 
                                                      color='DarkBlue', label='setosa')

groups.get_group('versicolor').plot.scatter(x='petal_length', y='petal_width', 
                                                     color='DarkGreen', label='versicolor', ax=ax)

groups.get_group('virginica').plot.scatter(x='petal_length', y='petal_width', 
                                                     color='Red', label='virginica', ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots() 
ax = tips.plot.scatter(x='total_bill', y='tip', ax=ax)
```

```python
fig, ax = plt.subplots() 
ax = tips.plot.hexbin(x='total_bill', y='tip', gridsize=10, ax=ax)
```

### 상자그래프 그리기

```python
fig, ax = plt.subplots() 
ax = tips.plot.box(ax=ax)
```
